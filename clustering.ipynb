```python
import spacy
spacy.load('es_core_web_md')
parser = spacy.es.Spanish()
import textacy
from textacy import preprocess
```

```python
salida = open('books', 'w')
```

```python
with open('sbwce/books/Books.en-es.es') as file:
    for line in file:
        line = preprocess.replace_urls(line)
        salida.write(preprocess.remove_punct(line))
```

```python
file = open('books')
raw = file.read()
doc = parser(raw)
```

```python
vocabulary = open('booksvocab', 'w')
from spacy.attrs import ORTH
counts = doc.count_by(ORTH)
```

```python
for word_id, count in sorted(counts.items(), reverse=True, key=lambda item: item[1]):
    if word_id >= 10 and count > 2:
        vocabulary.write("{} {}\n".format(doc.vocab.strings[word_id], count))

# Luego de esto, se puede utilizar todo para correr ./exchange
# ./exchange -c 40 -a 1000 -m 8000 --vocabulary=booksvocabstop -t 4 books books_withvocabstop_c40_a1000_m8000
```
